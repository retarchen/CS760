```python
data = pd.read_csv('hw3Data/emails.csv')
X = data.iloc[:, 1:-1]  
y = data['Prediction']
```

```python
n_splits = 5
fold_size = len(X) // n_splits

accuracy_scores = []
precision_scores = []
recall_scores = []

for fold in range(n_splits):
    test_indices = list(range(fold * fold_size, (fold + 1) * fold_size))
    train_indices = [i for i in range(len(X)) if i not in test_indices]

    X_train, X_test = X.values[train_indices], X.values[test_indices]
    y_train, y_test = y.values[train_indices], y.values[test_indices]

    fold_predictions = np.zeros(len(X_test))

    for i, test_instance in enumerate(X_test):
        a_=np.linalg.norm(X_train-X_test[i],axis=1)
        nearest_n = np.argmin(a_)
        fold_predictions[i] = y_train[nearest_n]

    accuracy = np.mean(y_test == fold_predictions)

    true_positives = np.sum((y_test == 1) & (fold_predictions == 1))
    predicted_positives = np.sum(fold_predictions == 1)
    precision = true_positives / predicted_positives if predicted_positives > 0 else 0.0

    actual_positives = np.sum(y_test == 1)
    recall = true_positives / actual_positives if actual_positives > 0 else 0.0


    accuracy_scores.append(accuracy)
    precision_scores.append(precision)
    recall_scores.append(recall)
    print(fold)

for fold in range(n_splits):
    print(f"Fold {fold + 1} - Accuracy: {accuracy_scores[fold]:.4f},"+
         f" Precision: {precision_scores[fold]:.4f}, Recall: {recall_scores[fold]:.4f}")
```

# Q2.3


```python
data = pd.read_csv('hw3Data/emails.csv')
X = data.iloc[:, 1:-1]  
y = data['Prediction']
```

```python

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def logistic_regression(X, theta):
    z = np.dot(X, theta)
    return sigmoid(z)

def cross_entropy_loss(y_hat, y):
    epsilon = 1e-15  # Small constant to avoid division by zero
    y_hat = np.maximum(epsilon, np.minimum(1 - epsilon, y_hat))
    return -np.mean(y * np.log(y_hat) + (1 - y) * np.log(1 - y_hat))

def gradient_descent(X, y, theta, learning_rate):
    y_hat = logistic_regression(X, theta)
    gradient = np.dot(X.T, (y_hat - y)) / len(y)
    return theta - learning_rate * gradient

learning_rate = 0.01
num_iterations = 1300
num_folds = 5
fold_size = len(X) // num_folds

accuracies = []
precisions = []
recalls = []

for fold in range(num_folds):
    start_idx = fold * fold_size
    end_idx = (fold + 1) * fold_size
    X_test = X.iloc[start_idx:end_idx, :]
    y_test = y.iloc[start_idx:end_idx]
    X_train = pd.concat([X.iloc[:start_idx, :], X.iloc[end_idx:, :]], axis=0)
    y_train = pd.concat([y.iloc[:start_idx], y.iloc[end_idx:]], axis=0)

    theta = np.zeros(X_train.shape[1])

    for i in range(num_iterations):
        theta = gradient_descent(X_train, y_train, theta, learning_rate)

    y_pred = logistic_regression(X_test, theta)
    y_pred_binary = (y_pred >= 0.5).astype(int)

    accuracy = np.mean(y_pred_binary == y_test)
    precision = np.sum((y_pred_binary == 1) & (y_test == 1)) / np.sum(y_pred_binary == 1)
    recall = np.sum((y_pred_binary == 1) & (y_test == 1)) / np.sum(y_test == 1)

    accuracies.append(accuracy)
    precisions.append(precision)
    recalls.append(recall)

    print(f"Fold {fold + 1} - Accuracy: {accuracy:.4f}, Precision: {precision:.4f}, Recall: {recall:.4f}")

avg_accuracy = np.mean(accuracies)
avg_precision = np.mean(precisions)
avg_recall = np.mean(recalls)
print(f"Average Accuracy: {avg_accuracy:.4f}, Average Precision: {avg_precision:.4f}, Average Recall: {avg_recall:.4f}")
```

# Q2.4

```python
data = pd.read_csv('hw3Data/emails.csv')
X = data.iloc[:, 1:-1]  
y = data['Prediction']
```

```python
n_splits = 5
fold_size = len(X) // n_splits

k_values = [1, 3, 5, 7, 10]

average_accuracies = []

for k in k_values:
    accuracy_scores = []

    for fold in range(n_splits):
        test_indices = list(range(fold * fold_size, (fold + 1) * fold_size))
        train_indices = [i for i in range(len(X)) if i not in test_indices]

        X_train, X_test = X.values[train_indices], X.values[test_indices]
        y_train, y_test = y.values[train_indices], y.values[test_indices]

        fold_predictions = np.zeros(len(X_test))

        for i, test_instance in enumerate(X_test):
            distances = np.linalg.norm(X_train - X_test[i], axis=1)
            nearest_indices = np.argsort(distances)[:k]
            nearest_neighbors = y_train[nearest_indices]
            fold_predictions[i] = np.bincount(nearest_neighbors).argmax()

        accuracy = np.mean(y_test == fold_predictions)
        accuracy_scores.append(accuracy)

    average_accuracy = np.mean(accuracy_scores)
    average_accuracies.append(average_accuracy)

    print(f"Avg. Accuracy (k={k}): {average_accuracy:.4f}")
```

```python

plt.plot(k_values, average_accuracies, marker='o')
plt.xlabel('k')
plt.ylabel('Average Accuracy')
plt.title('KNN 5-fold ')

```
